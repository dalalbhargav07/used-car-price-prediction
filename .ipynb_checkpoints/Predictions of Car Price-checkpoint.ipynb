## 1. Data collection

```python
# data colelction and preprocessing
import pandas as pd
import csv
from bs4 import BeautifulSoup
import requests
# for data visualisation and statistical analysis
import numpy as np
from sklearn.cross_validation import train_test_split as tts
import seaborn as sns
sns.set_style("darkgrid")
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from pylab import rcParams
from sklearn.metrics import r2_score
%matplotlib inline
```

```python
def get_ads_urls():
    urls_list = []
    # define the basic url to crawl on
    basic_url = "https://www.avito.ma/fr/maroc/voitures-à_vendre?mpr=500000000&o="
    # loop over the different urls for different ads
    for i in range(1,200):
        # get the page url
        url = basic_url+str(i)
        # get the request response
        r  = requests.get(url)
        data = r.text
        # transform it to beautifulsoap object
        soup = BeautifulSoup(data, "lxml")
        # loop over page links
        for div in soup.findAll('div', {'class': 'item-img'}):
            a = div.findAll('a')[0]
            urls_list.append(a.get('href'))


    df = pd.DataFrame(data={"url": urls_list})
    df.to_csv("./data/ads_urls.csv", sep=',',index=False)

# get the ads urls and save them in a file
get_ads_urls()
```

```python
def scraping_data(ad_url):
    r = requests.get(ad_url)
    data = r.text
    soup = BeautifulSoup(data, "html.parser")
    target_component = soup.findAll("h2",  {"class": ["font-normal", "fs12", "no-margin", "ln22"]})
    # create a list that will hold our component data
    results = []
    for i in target_component:
        results.append(''.join(i.findAll(text=True)).replace('\n',''))
    return results

def writing_data_to_csv(data):
    with open("./data/final.csv", "w") as f:
        writer = csv.writer(f)
        writer.writerows(data)
```

```python
# read the saved urls file as a dataframe 
urls = pd.read_csv("./data/ads_urls.csv")
# create  a list that will hold the final data
final_result = []
i = 1
# loop over the dataframe
for index, row in urls.iterrows():
    final_result.append(scraping_data(row['url']))
   
print('Scrapping the data finished')
# now that we have all the data we can write it in a csv file
writing_data_to_csv(final_result)
```

## 2. Data cleaning and pre-processing

Finally, now we got the raw data from the urls, this raw data can now be used for exploratary analysis as well as for data modelling after cleaning and pre-processing it further.

```python
# setting the column names
colnames=['price', 'year_model', 'toal_kms_car_ran', 'fuel_type', 'car_company', 'model', 'fiscal_power', 'sector', 'type', 'city'] 
# read the csv file
df = pd.read_csv("./data/final.csv", sep=",", names=colnames, encoding = "latin" ,header=None)
# let us check how our data looks as of now
df.head()
```

The very first thing that came to mind is to get rid of extra information such as all DH from the price, Annee Modele, other similart words and to also drop some unwanted column such as sector, type, etc.

### Cleaning and pre-processing each column

#### Price

The DH was removed from the price column. But, after exploring the data I found that there were some ads which were published without any value of price, hence those columns whose price was null were dropped.

```python
# dropping the rows which consist of null value in price
df = df[df.price.str.contains("DH") == True]
# remove the 'DH' caracters from the price
df.price = df.price.map(lambda x: x.rstrip('DH'))
df.price = df.price.str.replace(" ","")
# change it to integer value
df.price = pd.to_numeric(df.price, errors = 'coerce', downcast= 'integer')
```

#### year_model

```python
# remove thos rows doesn't contain the year_model value
df = df[df.year_model.str.contains("Année-Modèle") == True]
# remove the 'Année-Modèle:' from the year_model
df.year_model = df.year_model.map(lambda x: x.lstrip('Année-Modèle:').rstrip('ou plus ancien'))
# remove those lines having the year_model not set
df = df[df.year_model != ' -']
df = df[df.year_model != '']
# change it to integer value
df.year_model = pd.to_numeric(df.year_model, errors = 'coerce', downcast = 'integer')
```

#### toal_kms_car_ran

This column values consist of range of kms. Thus, instead of one hot encoding the data into different categories. I decided to take mean of the range values.

```python
# dropping the null values if any
df = df[df.toal_kms_car_ran.str.contains("Kilométrage") == True]
# remove the 'Kilométrage:' string from the mileage feature 
df.toal_kms_car_ran = df.toal_kms_car_ran.map(lambda x: x.lstrip('Kilométrage:'))
df.toal_kms_car_ran = df.toal_kms_car_ran.map(lambda x: x.lstrip('Plus de '))
# remove those lines having the mileage values null or '-'
df = df[df.toal_kms_car_ran != '-']
# we have only one value type that is equal to 500 000, all the other ones contain two values
if any(df.toal_kms_car_ran != '500 000'):
    # create two columns minKm and maxKm to calculate the mileage mean
    df['minKm'], df['maxKm'] = df.toal_kms_car_ran.str.split('-', 1).str
    # remove spaces from the maxKm & minKm values 
    df['maxKm'] = df.maxKm.str.replace(" ","")
    df['minKm'] = df.minKm.str.replace(" ","")
    df['maxKm'] = df['maxKm'].replace(np.nan, 500000)
    # calculate the mean of mileage 
    df.toal_kms_car_ran = df.apply(lambda row: (int(row.minKm) + int(row.maxKm)) / 2, axis=1)
    # now that the mileage is calculated so we do not need the minKm and maxKm values anymore
    df = df.drop(columns=['minKm', 'maxKm'])
```

#### Fuel Type

```python
# removing the 'Type de carburant:' string
df.fuel_type = df.fuel_type.map(lambda x: x.lstrip('Type de carburant:'))
```

#### Car Company & Model

```python
# removing the 'Marque:' string
df['car_company'] = df['car_company'].map(lambda x: x.replace('Marque:', ''))
df = df[df.car_company != '-']
# removing the 'Modèle:' string
df['model'] = df['model'].map(lambda x: x.replace('Modèle:', ''))
```

#### Fiscal Power

Uptil now, we have been dropping that whole row if there are null values. But, for this column there are many rows which consist of null values and we cannot afford to lose that muuch data as well. Hence, I decided to replace the null values with the mean of the column.

```python
df.fiscal_power.value_counts()
```

```python
# removing the 'Puissance fiscale:' string
df.fiscal_power = df.fiscal_power.map(lambda x: x.lstrip('Puissance fiscale:Plus de').rstrip(' CV'))
# replace the - with NaN values and convert them to integer values
df.fiscal_power = df.fiscal_power.str.replace("-","0")
# convert all fiscal_power values to numerical ones 
df.fiscal_power = pd.to_numeric(df.fiscal_power, errors = 'coerce', downcast= 'integer')
# now we need to fill those 0 values with the mean of all fiscal_power columns
df.fiscal_power = df.fiscal_power.map( lambda x : df.fiscal_power.mean() if x == 0 else x )
```

#### Fuel Type

```python
# removing those lines having the fuel_type not set
df = df[df.fuel_type != '-']
```

#### Dropping unwanted columns

```python
df = df.drop(columns=['sector', 'type'])
df = df[df.price <= 3000000]
```

## 4. Data Modelling & Evaluating the Model

### Dealing with categorical feature

In this dataset there 2 categorical features which are fuel_type, and car manufacturing company. The aim of this section is to pre process those features in order to convert them into numerical values so that they will be used to create a robust model which can predict the price accurately.
In general there are two ways for dealing with categorical variable, the first one is label encoding, and the second one is the one hot encoding, for our use case we will use the one hot encoding and the reason why I will choose this kind of data labeling is because I do not need any kind of data normalisation later, cause in sklearn packages methods used for regression first normalize the data and then work on regression model steps.

```python
## create a dataframefor testing
data = df[df.price < 400000]
X = data[['year_model', 'toal_kms_car_ran', 'fuel_type', 'fiscal_power', 'car_company']]
Y = data.price
X = pd.get_dummies(data=X)
# Avoiding Dummy Variable Trap
#X = X.drop(['fuel_type_Electrique', 'car_company_Zotye'],axis=1)

X.head()
```

### Data Splitting

Ideally, we split the datasets in three parts: Training set, Validation set and Test set. But,for our case study, will be dividing the dataset by 80% training set and 20% test set size.

```python
X_train, X_test, Y_train, Y_test = tts(X, Y, test_size = .20, random_state = 22)
```

### 4.1 Decision Tree Regression

```python
from sklearn.tree import DecisionTreeRegressor

dtr = DecisionTreeRegressor(max_features='auto')
dtr.fit(X_train, Y_train)
predicted = dtr.predict(X_test)
residual = Y_test - predicted

fig = plt.figure(figsize=(30,15))
ax1 = plt.subplot(121)
sns.distplot(residual, color ='grey')
plt.tick_params(axis='both', which='major', labelsize=20)
plt.title('Residual counts',fontsize=35)
plt.xlabel('Residual',fontsize=25)
plt.ylabel('Count',fontsize=25)

ax2 = plt.subplot(122)
plt.scatter(predicted, residual, color ='grey')
plt.tick_params(axis='both', which='major', labelsize=20)
plt.xlabel('Predicted',fontsize=25)
plt.ylabel('Residual',fontsize=25)
plt.axhline(y=0)
plt.title('Residual vs. Predicted',fontsize=35)

plt.show()

from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(Y_test, predicted))
print('RMSE:')
print(rmse)
```

```python
print('Variance score: %.2f' % r2_score(Y_test, predicted))
```

The root-mean-square deviation (RMSD) or root-mean-square error (RMSE) (or sometimes root-mean-squared error) is a frequently used measure of the differences between values (sample and population values) predicted by a model or an estimator and the values actually observed. The RMSD represents the sample standard deviation of the differences between predicted values and observed values. These individual differences are called residuals when the calculations are performed over the data sample that was used for estimation, and are called prediction errors when computed out-of-sample. The RMSD serves to aggregate the magnitudes of the errors in predictions for various times into a single measure of predictive power. RMSD is a measure of accuracy, to compare forecasting errors of different models for a particular data and not between datasets, as it is scale-dependent.

Variance is used to measure, how well your model or estimator will perform on unseen samples (test set).

** Interpretation of plots: **

It is evident from the plot that, our predicted price is closer to actual price because most of the points in "Residual vs Predicted" are near to zero, 

### 4.2 Linear Regression

```python
from sklearn import linear_model

regr = linear_model.LinearRegression()
regr.fit(X_train, Y_train)

predicted = regr.predict(X_test)
residual = Y_test - predicted

fig = plt.figure(figsize=(45,15))
ax1 = plt.subplot(121)
sns.distplot(residual, color ='teal')
plt.tick_params(axis='both', which='major', labelsize=20)
plt.title('Residual counts',fontsize=35)
plt.xlabel('Residual',fontsize=25)
plt.ylabel('Count',fontsize=25)

ax2 = plt.subplot(122)
plt.scatter(predicted, residual, color ='teal')
plt.tick_params(axis='both', which='major', labelsize=20)
plt.xlabel('Predicted',fontsize=25)
plt.ylabel('Residual',fontsize=25)
plt.axhline(y=0)
plt.title('Residual vs. Predicted',fontsize=35)

plt.show()

from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(Y_test, predicted))
print('RMSE:')
print(rmse)
```

```python
print('Variance score: %.2f' % r2_score(Y_test, predicted))
```

As can be seen from the figures and RMSE,variance values that linear regression has shown some improvement as compared to Decision Tree Regression. However, the model performance is yet not up to the par. RMSE is still high and also variance score is not remarkable.

### 4.3 Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import cross_val_score

r_sq = []
deep = []
mean_scores = []

for n in range(3, 11):
    gbr = GradientBoostingRegressor(loss ='ls', max_depth=n)
    gbr.fit (X, Y)
    deep.append(n)
    r_sq.append(gbr.score(X, Y))
    mean_scores.append(cross_val_score(gbr, X, Y, cv=10).mean())

plt_gbr = pd.DataFrame()

plt_gbr['mean_scores'] = mean_scores
plt_gbr['depth'] = deep
plt_gbr['R²'] = r_sq

f, ax = plt.subplots(figsize=(15, 5))

sns.barplot(data = plt_gbr, x='depth', y='R²')
plt.show()

f, ax = plt.subplots(figsize=(15, 5))
sns.barplot(data = plt_gbr, x='depth', y='mean_scores')
plt.show()
```

```python
gbr = GradientBoostingRegressor(loss ='ls', max_depth=5)
gbr.fit (X_train, Y_train)
predicted = gbr.predict(X_test)
residual = Y_test - predicted

fig = plt.figure(figsize=(30,30))
ax1 = plt.subplot(211)
sns.distplot(residual, color ='teal')
plt.tick_params(axis='both', which='major', labelsize=20)
plt.title('Residual counts',fontsize=35)
plt.xlabel('Residual',fontsize=25)
plt.ylabel('Count',fontsize=25)

ax2 = plt.subplot(212)
plt.scatter(predicted, residual, color ='teal')
plt.tick_params(axis='both', which='major', labelsize=20)
plt.xlabel('Predicted',fontsize=25)
plt.ylabel('Residual',fontsize=25)
plt.axhline(y=0)
plt.title('Residual vs. Predicted',fontsize=35)

plt.show()

rmse = np.sqrt(mean_squared_error(Y_test, predicted))
scores = cross_val_score(gbr, X, Y, cv=10)

print('\nCross Validation Scores:')
print(scores)
print('\nMean Score:')
print(scores.mean())
print('\nRMSE:')
print(rmse)
```

```python
print('Variance score: %.2f' % r2_score(Y_test, predicted))
```

As Gradient Boosting regressor is the winner, we will choose this model as our final model.
